# 1. Read the dataset to the python environment.

# 3. Plot a bar graph of male vs female buyers participated in the sales.

```python
# set width of bar
barWidth = 0.35
fig = plt.subplots(figsize =(12,8))
n=10
sale_series=[]
sale_min=car_data["Sale Price"].min()
sale_max=car_data["Sale Price"].max()
sale_ave=(sale_max-sale_min)/n

for i in range(0,n-1):
    label = str(round(sale_min)) + "-"
    sale_min=(round(sale_min+sale_ave))
    label += str(sale_min)
    sale_series.append(label)
#    print(label)

male_series=[]
female_series = []
for j in sale_series:
    sale_min,sale_max=j.split('-')
    gender_value= car_data[(car_data['Buyer Gender']=='Male') & (car_data['Sale Price'] > float(sale_min)) & (car_data['Sale Price'] < float(sale_max ))]
    male_series.append(gender_value.shape[0])

    # female count
    gender_value= car_data[(car_data['Buyer Gender']=='Female') & (car_data['Sale Price'] > float(sale_min)) & (car_data['Sale Price'] < float(sale_max ))]
    female_series.append(gender_value.shape[0])

# Set position of bar on X axis
br1 = np.arange(len(male_series))
br2 = [x + barWidth for x in br1]
 
# Make the plot
plt.bar(br1, male_series, color ='r', width = barWidth,
        edgecolor ='grey', label ='Male')
plt.bar(br2, female_series, color ='g', width = barWidth,
        edgecolor ='grey', label ='Female')
 
# Adding Xticks
plt.xlabel('Sale Price', fontweight ='bold', fontsize = 17)
plt.ylabel('Buyer Gender', fontweight ='bold', fontsize = 17)
#plt.xticks([r + barWidth for r in range(len(Male))],sale_series)
plt.xticks([r + barWidth for r in range(len(male_series))],sale_series)  
plt.legend()
plt.show()
```

# insight

    2.According to the sale price both smallest and largest values of sale price have more female buyers.
